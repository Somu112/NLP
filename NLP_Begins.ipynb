```python
from spacy.matcher import Matcher
```

```python
matcher = Matcher(nlp.vocab)
```

```python
#solar power
pattern_1 = [{'LOWER': 'solarpower'}]
pattern_2 = [{'LOWER': 'solar'},{'LOWER':'power'}]
pattern_3 = [{'LOWER': 'solar'},{'IS_PUNCT':True},{'LOWER':'power'}]
```

```python
matcher.add('SolarPower',[pattern_1,pattern_2,pattern_3])
```

```python
# Let's take a sample text
doc = nlp("The solar power industry continues to grow like anything \
since the demands for solarpower is increasing day by day!\
Solar-power driven cars are also in great demand!")
```

```python
found_match = matcher(doc)
```

```python
print(found_match)
```

```python
for match_id, start, end in found_match:
  string_id = nlp.vocab.strings[match_id]
  span = doc[start:end]
  print(start, end, span.text)
```

```python
pattern_1 = [{'LOWER': 'solarpower'}]
pattern_2 = [{'LOWER': 'solar'}, {'IS_PUNCT':True},{'LEMMA':'power'}]
```

```python
matcher.remove('SolarPower')
matcher.add('SolarPower2',[pattern_1, pattern_2])
```

```python
# Let's try to find the patterns in a different text. 
doc2 = nlp("Solar-Powered energy runs all our solar-powered cars!!!")
```

```python
found_match = matcher(doc2)
print(found_match)
```

## Lifecycle of NLP Project

1. Data Acquisition
2. Text cleaning
3. Preprocessing
4. Feature Engineering
5. Modeling
6. Evaluation
7. Deployment
8. Monitoring and Model Updating

**Overview**

- The first step in the process of developing any NLP system is to collect the data relevant to the given task. Even if we are building a Rule based system, we will still need some data to design and test our rules. This is by far the most intruiging part to deal with, since we have plethora of data available as of date on the internet and it is very hard to find good quality and relevant data from the available sources. Nevertheless, while we are working in the context of any given setup (lab., company, university, etc.) it is relatively easier, because now we just need to concentrate on the so called tools which we need to collect and read the data. 

- The data we get is seldom clean, and this is where text cleaning comes into play. By far, in the entire lifecycle, text cleaning is the most time consuming part of any delivery. 

- After we are done with cleaning, text data often has a lot of variations and needs to be converted into some canonical form. This part of action is the Text preprcessing part. We will discuss some important methods of cleaning and numerically representing the data when it's available as a text. (Count Vectorizer, TF-IDF, Frequency Distribution, Bag of Words, Word2Vec, Avg Word2Vec, etc.)

- Now we can further perform Feature engineering where we can find out or carve out indicators which are most suitable for the task in hand. We will discuss in depth some of the important and regularly used indicators like NER and POS. 

- Finally we will perform the required supervised or unsupervised learning methods (Naive Bayes, SVM, KMeans, Topic Modelling, etc.) and further we will evaluate how good or how bad our model is. 

- Once we are ready with the final version of our model, we will perform the deployment of the model. We will cover an in-depth use case/study of one of the methods by means of which we can deploy our NLP applications as web services. However, there are many other which we need to keep in mind in place, like Allen NLP, Hugging Face, Google Cloud NLP APIs, Microsoft Azure APIs, etc. 

- At the end of the day, we will have to constantly monitor our application which is deployed in Production (and is probably being used) and keep on improving the model with model re-training. The latter is where at any given scheduled period of time in any week/month, we take fresh data, prepare it in the same way as we have done with our Training data, and retrain the model with previous and/or new data. (suggestion: please make a habit of creating and documenting reusable data cleaning scripts!!!) 

## Important Tasks In Data Preprocessing

**Tokenization**

- This is by far the most basic operation we perform on any textual data, I.e., to break down any given doc. into tokens. 
- Using these tokens only, as we'll come to know during the course of our learning, that we would be able to further represent our data as well as train ML/DL Models on top of it. 
- Even special characters are valid tokens as part of the doc we have in hand, and depending on the use case we are dealing with, we may have to treat/remove/replace/etc. such special characters. (In this context, we will talk about a concept called Stop Words!!)

**Parts of Speech**

- In ay given language we have a set of 8-10 at least the basic parts of speech , like we have 8 of them in English! 
- SPACY or NLTK or similar libraries provide the capability to link each token, inclusive of the special characters, to its relevant POS (Parts of Speech) Tag.
- This POS Tagging activity is a very important activity when it comes to Text mining, where one would like to understand the composition of the statements which are being used in a given document. 
- E.g., if we are going to create an NLP application where we upload any document and the application will tell us the composition of the language which is used within the document, then we need such POS Tagging functionalities!

**Entities and Labels --> NER**

- Entities can be thought of proper nouns which are nothing but names of identities of known objects/places/companies/etc.
- For example, Apple, India, Money, Country, City, etc. 
- If we are able to detect the entities from any given text and are also able to label it (or understand it) then we are essentially solving a problem of **Named Entity Recognition**.
- NER is by far the most interesting of the use cases in NLP applications as a start. 
- E.g., if we'd like to create an NLP application where we upload one/set of document(s), so that the application can identify to which company or to which topic those documents belong to, then it would help in the indexing of those documents. Somewhat similar approach is used by any search engine like Google when they use *PageRank* algorithm to index and label the websites.
